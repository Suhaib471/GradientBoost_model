## What is the account roll forward rate for the months of Mar , Apr, May and June for TB0?

```python
data_file.keys()
```

```python
df = data_file
```

```python
# start_data = df['Loan Details']
mar1_data = df['Loan_Status_MartoMay'].loc[df['Loan_Status_MartoMay']['Month'] == 'March']
apr1_data = df['Loan_Status_MartoMay'].loc[df['Loan_Status_MartoMay']['Month'] == 'April']
may1_data = df['Loan_Status_MartoMay'].loc[df['Loan_Status_MartoMay']['Month'] == 'May']
apr2_data = df['Loan_Status_AprtoJun'].loc[df['Loan_Status_AprtoJun']['Month'] == 'April']
may2_data = df['Loan_Status_AprtoJun'].loc[df['Loan_Status_AprtoJun']['Month'] == 'May']
jun2_data = df['Loan_Status_AprtoJun'].loc[df['Loan_Status_AprtoJun']['Month'] == 'June']
```

```python
apr2_data.head()
```

```python
mar_data = pd.merge(mar1_data, apr2_data, how='left', on='Loan_id', suffixes=('_mar01', '_mar31'))
apr_data = pd.merge(apr1_data, may2_data, how='left', on='Loan_id', suffixes=('_apr01', '_apr31'))
may_data = pd.merge(may1_data, jun2_data, how='left', on='Loan_id', suffixes=('_may01', '_may31'))
```

```python
data1 = pd.merge(mar_data, apr_data, how='outer', on='Loan_id')
data1 = pd.merge(data1, may_data, how='outer', on='Loan_id')
```

```python
print(mar_data.shape)
print(apr_data.shape)
print(may_data.shape)
print(data1.shape)
```

```python
data2['RB2'] = np.where((data2['Bucket_apr01']=='TB0')&(data2['Bucket_mar01']=='TB0'), 1, 0)
data2['RB3'] = np.where(((data2['Bucket_may01']=='Nan')&(data2['Bucket_apr01']=='TB0')
                         |(data2['Bucket_may01']=='REGULAR')&(data2['Bucket_apr01']=='TB0')
                        |(data2['Bucket_may01']=='TB0')&(data2['Bucket_apr01']=='TB1')
                        ), 1, 0)
data2['RB4'] = np.where(((data2['Bucket_may31']=='Nan')&(data2['Bucket_may01']=='TB0')
                         |(data2['Bucket_may31']=='REGULAR')&(data2['Bucket_may01']=='TB0')
                        |(data2['Bucket_may31']=='TB0')&(data2['Bucket_may01']=='TB1')
                        ), 1, 0)
```
